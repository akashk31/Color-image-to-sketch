```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.ndimage
```

```python
img=mpimg.imread("vi.jpg")
```

```python
plt.imshow(img)
```

```python
rgb_weights = [0.2989, 0.5870, 0.1140]
grayscale_image = np.dot(img[...,:3], rgb_weights)
plt.imshow(grayscale_image, cmap=plt.get_cmap("gray"))
```

```python
inverted_img=255-grayscale_image
plt.imshow(inverted_img)
```

```python
blur_img=scipy.ndimage.filters.gaussian_filter(inverted_img,sigma=5)
plt.imshow(blur_img)
```

```python
def dodge(front,back):
    result=front*255/(255-back)
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```python
final_img=dodge(blur_img,grayscale_image)
plt.figure(figsize=(8,6))
plt.imshow(img)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(final_img,cmap="gray")
```
